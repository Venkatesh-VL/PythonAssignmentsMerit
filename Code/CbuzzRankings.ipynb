```python
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
pd.set_option('display.max_rows',None)
```

```python
os.chdir(r"C:\Users\krthg\OneDrive\Documents\Merit\PythonAssignments")
```

```python
testbatrankings = []
odibatrankings = []
t20batrankings = []
testbowlrankings = []
odibowlrankings = []
t20bowlrankings = []
testallrankings = []
odiallrankings = []
t20allrankings = []
testteam = []
oditeam = []
t20team = []
```

```python
driverPath = r"C:\Users\krthg\OneDrive\Documents\Merit\PythonAssignments\Webdriver\chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverPath)
driver.get("https://www.cricbuzz.com/")
time.sleep(1)
driver.find_element_by_xpath("//div[@id='rankingDropDown']").click()
driver.find_element_by_xpath("//a[@id='batsmen-tab']").click()
Testrankings = driver.find_elements_by_xpath("//a[@id='batsmen-tests-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for i in Testrankings:
    if i.text != '':
        testbatrankings.append(i.text.split("\n"))
print(testbatrankings) 
driver.find_element_by_xpath("//a[@id='batsmen-tab']").click()
driver.find_element_by_xpath("//a[@id='batsmen-odis-tab']").click()
Odirankings = driver.find_elements_by_xpath("//a[@id='batsmen-odis-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for j in Odirankings:
    if j.text != '':
        odibatrankings.append(j.text.split("\n"))
print(odibatrankings)
driver.find_element_by_xpath("//a[@id='batsmen-tab']").click()
driver.find_element_by_xpath("//a[@id='batsmen-t20s-tab']").click()
T20rankings = driver.find_elements_by_xpath("//a[@id='batsmen-t20s-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for k in T20rankings:
    if k.text != '':
        t20batrankings.append(k.text.split("\n"))
print(t20batrankings)

# Bowling rankings

driver.find_element_by_xpath("//a[@id='bowlers-tab']").click()
driver.find_element_by_xpath("//a[@id='bowlers-tests-tab']").click()
Testbowlrankings = driver.find_elements_by_xpath("//a[@id='bowlers-tests-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for i in Testbowlrankings:
    if i.text != '':
        testbowlrankings.append(i.text.split("\n"))
print(testbowlrankings) 
driver.find_element_by_xpath("//a[@id='bowlers-tab']").click()
driver.find_element_by_xpath("//a[@id='bowlers-odis-tab']").click()
Odibowlrankings = driver.find_elements_by_xpath("//a[@id='bowlers-odis-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for j in Odibowlrankings:
    if j.text != '':
        odibowlrankings.append(j.text.split("\n"))
print(odibowlrankings)
driver.find_element_by_xpath("//a[@id='bowlers-tab']").click()
driver.find_element_by_xpath("//a[@id='bowlers-t20s-tab']").click()
T20bowlrankings = driver.find_elements_by_xpath("//a[@id='bowlers-t20s-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for k in T20bowlrankings:
    if k.text != '':
        t20bowlrankings.append(k.text.split("\n"))
print(t20bowlrankings)

# All-rounders rankings

driver.find_element_by_xpath("//a[@id='all-rounders-tab']").click()
driver.find_element_by_xpath("//a[@id='allrounders-tests-tab']").click()
Testallrankings = driver.find_elements_by_xpath("//a[@id='allrounders-tests-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for i in Testallrankings:
    if i.text != '':
        testallrankings.append(i.text.split("\n"))
print(testallrankings) 
driver.find_element_by_xpath("//a[@id='all-rounders-tab']").click()
driver.find_element_by_xpath("//a[@id='allrounders-odis-tab']").click()
Odiallrankings = driver.find_elements_by_xpath("//a[@id='allrounders-odis-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for j in Odiallrankings:
    if j.text != '':
        odiallrankings.append(j.text.split("\n"))
print(odiallrankings)
driver.find_element_by_xpath("//a[@id='all-rounders-tab']").click()
driver.find_element_by_xpath("//a[@id='allrounders-t20s-tab']").click()
T20allrankings = driver.find_elements_by_xpath("//a[@id='allrounders-t20s-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-lst-itm text-center')]")
for k in T20allrankings:
    if k.text != '':
        t20allrankings.append(k.text.split("\n"))
print(t20allrankings)

# Teams rankings

driver.find_element_by_xpath("//a[@id='teams-tab']").click()
driver.find_element_by_xpath("//a[@id='teams-tests-tab']").click()
Testteamrankings = driver.find_elements_by_xpath("//a[@id='teams-tests-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-brdr-thin-btm text-center')]")
for i in Testteamrankings:
    if i.text != '':
        testteam.append(i.text.split("\n"))
print(testteam) 
driver.find_element_by_xpath("//a[@id='teams-tab']").click()
driver.find_element_by_xpath("//a[@id='teams-odis-tab']").click()
Oditeamrankings = driver.find_elements_by_xpath("//a[@id='teams-odis-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-brdr-thin-btm text-center')]")
for j in Oditeamrankings:
    if j.text != '':
        oditeam.append(j.text.split("\n"))
print(oditeam)
driver.find_element_by_xpath("//a[@id='teams-tab']").click()
driver.find_element_by_xpath("//a[@id='teams-t20s-tab']").click()
T20teamrankings = driver.find_elements_by_xpath("//a[@id='teams-t20s-tab']//following::div[contains(@class,'cb-col cb-col-100 cb-font-14 cb-brdr-thin-btm text-center')]")
for k in T20teamrankings:
    if k.text != '':
        t20team.append(k.text.split("\n"))
print(t20team)


driver.close()
```
